```python
#linear regression
import matplotlib.pyplot as plt
import seaborn as sns
diamonds = sns.load_dataset('diamonds').dropna()
diamonds.head()
```

```python
diamonds.shape
```

```python
data = diamonds.sample(n=190, random_state=38)
```

```python
sns.set_style('whitegrid')
sns.regplot(data=data, x='carat',y='price')
plt.show()
```

```python
sns.regplot(data=data,x='carat',y='price',fit_reg=False)
```

```python
sns.regplot(data=diamonds,x='carat',y='price')#,scatter=False)
```

```python
sns.countplot(data=diamonds, x='cut')
```

```python
sns.regplot(data=data,y='price',x='carat',ci=None)
```

```python
cut_map = {
    'Fair':1,
    'Good':2,
    "Very Good":3,
    "Premium":4,
    "Ideal":5
}
```

```python
data['cut_value']=data.cut.map(cut_map).cat.as_ordered()
```

```python
sns.regplot(x='cut_value',y='price',data=data)
plt.show()
```

```python
import numpy as np
data['cut_value']=data.cut.map(cut_map).cat.as_ordered().astype(np.int8)
sns.regplot(x='cut_value',y='price',data=data,x_jitter=.1)
plt.show()
```

```python
sns.regplot(x='cut_value',y='price',data=data,x_estimator=np.mean)
plt.show()
```
