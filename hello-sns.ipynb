```python
delays=pd.read_csv('data/flight_delays.csv', index_col='Month')
delays.head()
```

```python
#Distributions
iris=pd.read_csv("iris/iris.csv")
iris.head()
```

```python
sns.displot(iris['petal_length'],kde=False)
```

```python
sns.kdeplot(data=iris['petal_length'],shade=True)
```

```python
#2d kdeplot
sns.jointplot(x=iris['petal_length'],y=iris['sepal_width'],kind='kde',shade=True)
```

```python
setosa = iris.loc[iris['species']=='Iris-setosa']
versicolor = iris.loc[iris['species']=='Iris-versicolor']
virginica = iris.loc[iris['species']=='Iris-virginica']
```

```python
versicolor.head()
```

```python
# histograam for each species
sns.displot(setosa['petal_length'],label="Iris-setosa",kde=False)
sns.displot(versicolor['petal_length'],label="Iris-versicolor",kde=False)
sns.displot(virginica['petal_length'],label="Iris-virginica",kde=False)

#plt.title("Histogram of Petal Lengths by Species")
plt.legend()
```

```python
# histograam for each species
sns.kdeplot(setosa['petal_length'],label="Iris-setosa",kde=True)
sns.kdeplot(versicolor['petal_length'],label="Iris-versicolor",kde=False)
sns.kdeplot(virginica['petal_length'],label="Iris-virginica",kde=False)

#plt.title("Histogram of Petal Lengths by Species")
plt.legend()
```
