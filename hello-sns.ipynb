```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup complete")
```

```python
fifa_path ="data/fifa.csv"

fifa = pd.read_csv(fifa_path, index_col = 'Date',parse_dates=True)
```

```python
fifa.head()
```

```python
#set width and height of the figure
plt.figure(figsize=(16,6))

#line chart showing how FIFA rankings evolved over time

sns.lineplot(data=fifa)
```

```python
delays=pd.read_csv('data/flight_delays.csv', index_col='Month')
delays.head()
```

```python
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights by Month")
sns.barplot(x=delays.index,y=delays['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

```python
#Heatmap
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for each Airline, By Month")
sns.heatmap(data=delays, annot=True)
plt.xlabel("Airline")
```
