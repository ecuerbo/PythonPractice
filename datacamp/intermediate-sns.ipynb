```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file = 'https://assets.datacamp.com/production/repositories/2210/datasets/794e0759b73a2d80baa5d8fb88636a47965139d3/college_datav3.csv'
df = pd.read_csv(file)
```

```python
for style in ['white', 'dark', 'whitegrid','darkgrid','ticks']:
    sns.set_style(style)
    sns.displot(df['Tuition'], color='g');
```
