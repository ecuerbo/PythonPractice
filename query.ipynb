```python
#from https://towardsdatascience.com/10-examples-that-will-make-you-use-pandas-query-function-more-often-a8fb3e9361cb
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.randint(10,size=(5,6)), columns=list("ABCDEF"))
df
```

```python
df.query("A >4")
```

```python
df.query("A >4 and E > 4")
```

```python
df.query("A > 2 or F > 4")
```

```python
df.query("not(A>2 or F>4)")
```

```python
df.query("A*2 < 5")
```

```python
df.F = -1*df.F
df.query("abs(F)>6")
```

```python
df.query("sqrt(A) > 2")
```

```python
df["G"] = ["Agnes", "Agatha", "Felicity", "Anastasia", "Cecilia"]
df.query("G > 'Agn'")
```

```python
df["Date"] = pd.date_range(start="2021-04-05", periods=5, freq='M')
df
```

```python
df.query("Date.dt.month > 6")
```

```python
df.query("A + B < C")
```
