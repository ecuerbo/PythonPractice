```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('pokemon.csv')

sns.catplot(x='Stage', y='Attack', data= df, kind='point')
sns.catplot(x='Stage', y='Defense', data = df, kind ='strip')
```

```python
sns.factorplot(x='Stage', y='Attack', data= df)
sns.factorplot(x='Stage', y='Defense', data = df)
```

```python
#violinplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('pokemon.csv')

sns.catplot(x='Type 1', y='Attack',
           kind='violin', data=df,h=2,aspect=2)

plt.show()
```

```python
#barplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('pokemon.csv')

sns.catplot(x='Type 1',y='Defense', kind='bar',
            hue='Stage',data=df, palette='pastel')

plt.show()
```

```python
#boxplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x='Stage', y='Defense', kind='box', data=df)
```

```python
#strip plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('pokemon.csv')
sns.catplot(x='Stage', y='Defense',
           kind='strip', data=df)
```

```python
sns.catplot(x="Stage", kind='count', data = df)
```
