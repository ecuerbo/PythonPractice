```python
import matplotlib.pyplot as plt
import seaborn as sns

x=['sun', 'mon','tue','wed','thu','fri','sat']
y=[5,6.7,4,6,2,4.9,1.8]

ax = sns.stripplot(x,y);

#giving labels to x-axix and y-axis
ax.set(xlabel='Days', ylabel='Amount_spend')

#giving title to the plot
plt.title('My First Graph');

#function to show plot
plt.show()
```

```python
#Stripplot using inbuilt data-set given in seaborn : 

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

iris = sns.load_dataset('iris')

sns.stripplot(x='species', y='sepal_length', data= iris, palette='colorblind')

plt.title('Graph')
plt.show()
```

```python
#swarmplot

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
iris = sns.load_dataset('iris')

sns.swarmplot(x='species',y='sepal_length', data=iris, palette='colorblind')

plt.title('Iris Petal Length by Species')
plt.show
```

```python
#barplot
import seaborn as seaborn

df = sns.load_dataset('tips')

sns.barplot(x='sex', y= 'total_bill', hue='smoker',data = df, palette='colorblind')
```

```python
#countplot
import seaborn as sns

df = sns.load_dataset('tips')
sns.countplot(x='sex',hue='smoker',data=df)
```

```python
#boxplot()
import seaborn as sns

df = sns.load_dataset('tips')

sns.boxplot(x='day',y='total_bill', data=df, hue='smoker', palette='coolwarm')
```

```python
#boxplot()
import seaborn as sns

df = sns.load_dataset('tips')

sns.boxplot(x='day',y='total_bill', data=df, hue='sex', palette='coolwarm')
```

**Violinplot**

It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution.

```python
import seaborn as sns

df = sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill', data=df,
              hue='sex', split=True, palette='colorblind')
```

```python
'''Stripplot
It basically creates a scatter plot based on the category.'''

import seaborn as sns
df = sns.load_dataset('tips')
sns.stripplot(x='day', y='total_bill', data=df,
             jitter=True,hue ='smoker', dodge=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=df,
              jitter=True, hue='sex',dodge=True)
```
